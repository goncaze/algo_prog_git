# **<font color=green>Estrutura de repetição: </font>**  
#### Uma estrutura de repetição, de uma forma simples e elegante, nos possibilita repetir a execução de (passos/comandos/códigos).  

#### <font color=red>__Problema:__</font> suponhamos que ainda não inventaram as estruturas de repetição, então, escreva um código que imprima na tela uma sequência de números de zero a três.

#### <font color=blue>__Possível solução:__</font>
```
print(0)
print(1)
print(2)
print(3)
```
#### Fácil, correto? Agora tente imprimir outra sequência de zero a trezentos ou de menos 300 a mais 300 ou, então, de zero a dez mil. Humnm, agora não está nada fácil, concorda? Podemos considerar que essa dificuldade é a motivação para usarmos estruturas de repetição.

#### Na maioria das liguagens de programação como o Python existem o __Para__ (__for__) e o __Enquanto__ (__while__) como estruturas de repetição. Veremos exemplos de uso dessas duas estruturas nas próximas células.
<br>

# **<font color=green>Para (for): </font>**  
#### Utilizando o problema anterior de imprimir uma sequência de zero a três, temos que o modelo da estrutura repetição do comando __para__ é o seguinte: <br>
   
> #### para ( valor_inicial; condição; valor_do_incremento )
> #### {
> #### &nbsp;&nbsp;&nbsp;&nbsp;bloco de comandos
> #### }

<br>

#### Sendo assim, temos que:
> #### __valor_inicial__ é uma variável que recebe o valor zero (var = 0)
> #### __condição__ é a variável menor igual a três (var <= 3)
> #### __valor_do_incremento__ é a adição mais um à variável (var = var + 1)
> #### __bloco de comandos__ é o comando imprimir(var).

<br>

#### Dessa maneira, temos a seguinte estrutura de repetição:  <br>  
> #### para ( __var = 0__; __var <= 3__; __var = var + 1__ )
> #### {
> #### &nbsp;&nbsp;&nbsp;&nbsp;__imprimir(var)__
> #### }

<br>

#### <font color=green>__Passos da execução:__</font>
> #### 1. A variável var é zero,
> #### 2. Enquanto variável var for menor ou igual a três,
> #### 3. Imprima o valor da variável var,
> #### 4. Em seguida, incremente o valor de variável var em mais um,
> #### 5. Finalmente, repita todos os passos anteriores.

<br>

#### __Observação:__ a estrutura de repetição somente será interrompida quando a condição de __var <= 3__ não for mais satisfeita, ou seja, a repetição será interrompida quando o valor da variável var for igual a quatro (__var = 4__).

#### Agora chega de teoria! Na próxima célula faremos o exemplo da estrutura de repetição usando a linguagem de programação Python.


# **<font color=green>for (para) em Python: </font>**  

#### Para percorrer uma sequência numérica em Python temo a função __range(início, fim)__. A função range retorna uma sequência de números que, por padrão, inicia em zero. Mas, atenção: a função range não retorna o último número, isto é, se você deseja um intervalo até o número três, então use o número quatro. Exemplo:  <br>

>    #### range(0, 4)

#### Tem como resultado um intervalo que vai de zero a três.  
#### Sendo assim, podemos imprimir uma sequência de números em Python da seguinte forma:  <br>

> #### for var in range(0, 4):  
> #### &nbsp;&nbsp;&nbsp;&nbsp;print(var)

<br>

#### __Leitura do código:__ Para var em um intervalo de zero a três, imprima o valor de var.  
<br>

````python
# Executando o código em Python
for var in range(0, 4):
    print(var)
````

# **<font color=green>Enquanto (while): </font>**  
#### Utilizando o primeiro problema de imprimir uma sequência de zero a três, temos que o modelo da estrutura repetição do comando __enquanto__ é o seguinte: <br>
   
> #### enquanto( condição )
> #### {
> #### &nbsp;&nbsp;&nbsp;&nbsp;bloco de comandos
> #### }

<br>

#### Sendo assim, temos que:
> #### __condição__ quando satisfeita interrompe a repetição e enquanto não é satisfeita permanece repetindo o bloco de comandos;
> #### __bloco_de_comandos__ instruções que devem ser repetidas.

<br>

#### Dessa maneira, temos a seguinte estrutura de repetição:  <br>  
> #### __var__ = 0
> #### enquanto( __var <= 3__ )
> #### {
> #### &nbsp;&nbsp;&nbsp;&nbsp;__imprimir(var)__
> #### &nbsp;&nbsp;&nbsp;&nbsp;__var__ = var + 1
> #### }

<br>

#### <font color=green>__Passos da execução:__</font>
> #### 1. Uma variável __var__ iniciada com valor zero (var = 0),
> #### 2. __Enquanto__ variável var for menor ou igual a três (var <= 3),
> #### 3. Imprima o valor da variável var __imprimir(var)__,
> #### 4. Em seguida, __incremente__ o valor de var em mais um (__var__ = var + 1),
> #### 5. Finalmente, repita todos os passos anteriores.

<br>

#### __Observação 1:__ a variável __var__ é criada apenas uma única vez e isso ocorre quando lhe é atribuído o valor zero (var = 0). A sua "criação" não faz parte da estrutura de repetição. Porém, a reatribuição de valores faz parte da repetição.
#### __Observação 2:__ a estrutura de repetição somente será interrompida quando a condição de __var <= 3__ não for mais satisfeita, ou seja, a repetição será interrompida quando o valor da variável var for igual a quatro (__var = 4__).

#### Agora chega de teoria! Na próxima célula faremos o exemplo da estrutura de repetição usando a linguagem de programação Python.


# **<font color=green>while (enquanto) em Python: </font>**  

#### Para percorrer uma sequência numérica em Python com a estrutura de repetição while, precisaremos declarar uma variável com um valor inicial.
#### Sendo assim, podemos imprimir uma sequência de números em Python da seguinte forma:  <br>

> #### var = 0
> #### while(var <= 3):  
> #### &nbsp;&nbsp;&nbsp;&nbsp;print(var)
> #### &nbsp;&nbsp;&nbsp;&nbsp;var = var + 1

<br>

#### __Leitura do código:__ enquanto __var__, que inicialmente recebeu valor zero, for menor ou igual a três, imprima o valor de var. Em seguida, __incremente__ o valor de var em __mais um__. Por fim, repita todos os passos.


````python
# Executando o código em Python
var = 0
while(var <= 3):
    print(var)
    var = var + 1
````
