## Conversão de tipos

Podemos converter os tipos das variáveis e tal processo também conhecido como casting.

int() => converte um valor para inteiro.</br>
str() => converte um valor para string.</br>
float() => converte um valor para float.</br>

```python
int("8") # converte de texto "8" para um valor de inteiro 8.

str(8) # faz o oposto do exemplo acima.

float("2.5") # converte de texto "2.5" para um valor de ponto flutuante 2.5

print(f"int = {int('8')} || float = {float('2.5')}")
```

## Operadores aritméticos
Operadores aritméticos são usados ​​com valores numéricos para realizar operações matemáticas comuns:
+ soma (+),
- subtração (-),
* multiplicacão (*),
+ divisão (/).
+ módulo (%),
+ exponenciação (**)
+ resto da divisão (//)

```python
# Somar dois números
resultado = 5 + 3

print("O resultado da soma é: ", resultado)
```

```python
# Subtrair dois números
resultado = 3 - 5

print("O resultado da subtração é: ", resultado)
```

```python
# Multiplicar dois números
resultado = 3 * 5

print("O resultado da multiplicação é: ", resultado)
```

```python
# Dividir dois números
resultado = 5 / 3

print("O resultado da divisão é: ", round(resultado, 2))
```

```python
# Módulo
resultado = 5 % 3

print("O módulo é: ", resultado)
```

```python
# Exponenciação
resultado = 3 + 5

print("O resultado da Exponenciação é: ", resultado)
```

```python
# Resto da Divisão
resultado = 5 // 3

print("O resultado da Floor division é: ", resultado)
```

## Operadores de comparação
Operadores de comparação são usados ​​para comparar dois valores

```python
x = 3
y = 5

# Igual a
print(x == y)
```

```python
# Não igual
print(x != y)
```

```python
# Maior que
print(x > y)
```

```python
# Menor que
print(x < y)
```

```python
# Maior que ou igual a
print(x >= y)
```

```python
# Menor que ou igual a
print(x <= y)
```

```python
x = input("Digite um valor: ")
y = input("Digite outro valor")

x = float(x)
y = float(y)

igual = x == y
maior = x > y
diferente = x != y

mensagem = f"Igual = {igual} || Maior = {maior}"
mensagem += f" || Diferente {diferente}"

print(mensagem)
```

## Operadores lógicos

Operadores lógicos são usados ​​para combinar instruções condicionais.

```python
x = 5
```

```python
# (x é igual a 3?) E (x é menor que ou igual a 5?)
print(x == 3 and x >= 5)
```

```python
# (x é igual a 3?) OU (x é maior que ou igual a 5?)
print(x == 3 or x >= 5)
```

```python
# NÃO ((x é igual a 3?) E (x é maior que ou igual a 5?))
print(not(x == 3 and x >= 5))
```

```python
# NÃO ((x é igual a 3?) OU (x é maior que ou igual a 5?))
print(not(x == 3 or x >= 5))
```

## Operadores de identidade

Os operadores de identidade são usados ​​para comparar os objetos, não se forem iguais, mas se forem realmente o mesmo objeto, com a mesma localização de memória.

```python
x = 5
y = 20
z = 10
```

```python
# x é y?
print(x is y)
```

```python
# x é z?
print(x is z)
```

```python
# x não é y?
print(x is not y)
```

```python
print("x => ", hex(id(x)))
print("z => ", hex(id(z)))
print("y => ", hex(id(y)))
```
