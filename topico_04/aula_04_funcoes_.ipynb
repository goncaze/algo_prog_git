# **<font color=green>Funções: </font>**  


#### Uma função é como um bloco de código que no mínimo possui um nome e somente é executado quando é chamado/invocado. Em outras palavras, uma função nunca será executada enquanto não for chamada.  

#### Outra característica interessante é a de que esse bloco de códigos (função) deve apresentar a mesma saída de dados/informação para a mesma entrada de dados/informação.

#### Sendo assim, imagine uma função para somar dois números cujo nome seja somar(a, b). Nas várias chamadas da função somar(a, b) entramos sempre com os mesmos números a=2 e b=3, então, a função sempre deverá ter como saída/retorno a resposta 5. Isto implica dizer que somar(2, 3) é igual a 5 sempre.

#### Mas, como já deve ter percebido além do bloco de código, uma função não necessariamente possui apenas o nome dela própria, podendo ter o que é denominado de parâmetros. As variáveis __a__ e __b__ são parâmetros da função somar(a, b).

#### Os parâmetros permitem que dados sejam passados para funções. Perceba que isto é muito útil pois imagine que uma função somar() sem parâmetros teria dificuldade ou seria imposível nos dar a resposta para quaisquer valores que as variáveis __a__ e __b__ pudessem assumir. Em outras palavras, como a função somar() poderia determinar quais valores queremos que ela calcule?

## <font color=green>__Criar a função somar(a,b) em Python__</font>

#### A partir de agora iremos criar funções e em seguida invocá-las para execução

```python
# Função somar

def somar(a, b):
  resposta = a + b
  return resposta
```

```python
# Agora, vamos chamar a função para somar 2 e 3

somar(2, 3)
```

```python
# Agora, vamos guardar o retorno da função somar em uma variável resultado e imprimir essa variável

resultado = somar(2, 3)

print("O resultado é: ", resultado)
```

```python
# Com mais DETALHE iremos recriar a Função somar

def somar(a: float, b: float) -> float:
  resposta: float = a + b
  return resposta
```

```python
resultado = somar(10, 100)

print(f"{resultado = }")
```

# **<font color=green>Parâmetro com valor padrão: </font>**  

#### Algo interessante de Python é que podemos definir valores default/padrão para os parâmetros da função. Isto quer dizer que, se não passarmos os valores dos parâmetros da função, ela própria ira assumir os valores definidos como o padrão/default. Observação: haverá casos em que esta capacidade será muito útil.
#### Vejamos um exemplo no diagrama a seguir.


 <br><br><br>

## <font color=green>__Com valores default recriar a função somar(a,b) em Python__</font>

```python
# Com valores default iremos recriar a Função somar

def somar(a: float = 3, b: float = 5) -> float:
  resposta: float = a + b
  return resposta
```

```python
resultado = somar(b = 8, a = 5)

print(f"{resultado = }")
```
